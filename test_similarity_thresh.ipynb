```python
ident1 = pd.read_csv('results/extract_dialogue/identifications/LFB1_full.csv')
ident2 = pd.read_csv('results/extract_dialogue/identifications/LFB1_full_2.csv')
```

```python
def get_hallucinations(ident1, ident2, verbose=False):
    ident1.replace(np.nan, '', inplace=True)
    ident2.replace(np.nan, '', inplace=True)
    
    hallucinations = np.zeros(len(ident1), dtype=bool)
    for i in range(len(ident1)):
        if ident1.iloc[i].transcription != ident2.iloc[i].transcription:
            hallucinations[i] = True
            if verbose:
                print(f"Ident 1: {ident1.iloc[i].transcription}")
                print(f"Ident 2: {ident2.iloc[i].transcription}")
        if ident1.iloc[i].transcription == '' and ident2.iloc[i].transcription == '':
            hallucinations[i] = True
            if verbose:
                print(f"Ident 1: {ident1.iloc[i].transcription}")
                print(f"Ident 2: {ident2.iloc[i].transcription}")
    return hallucinations

hallucinations = get_hallucinations(ident1, ident2)
hallucinations.sum(), len(hallucinations)
```

```python
from sklearn.metrics import precision_recall_fscore_support

def apply_similarity_thresh(df, similarity_thresh):
    dist_thresh = 1 - similarity_thresh
    df = df.copy()
    for i in range(len(df)):
        trainer_avg_dist = df.loc[i, 'trainer_dist']
        trainee_avg_dist = df.loc[i, 'trainee_dist']
        
        se_speaker = None
        if trainer_avg_dist < trainee_avg_dist and trainer_avg_dist < dist_thresh:
            se_speaker = 'trainer'
        elif trainee_avg_dist < trainer_avg_dist and trainee_avg_dist < dist_thresh:
            se_speaker = 'trainee'
        else:
            se_speaker = 'unknown'
        
        df.loc[i, 'speaker'] = se_speaker

    return df

def get_hallucination_metrics(ident, hallucinations):
    ident.replace(np.nan, '', inplace=True)

    pred_hallucinations = np.zeros(len(ident), dtype=bool)

    for i in range(len(ident)):
        if ident.iloc[i].speaker == 'unknown':
            pred_hallucinations[i] = True
    
    precision, recall, f1, _ = precision_recall_fscore_support(hallucinations, pred_hallucinations, average='binary')

    metrics = {
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'custom': 2*precision + recall/2
    }

    return metrics

results = pd.DataFrame(columns=['similarity_thresh', 'precision', 'recall', 'f1', 'custom'])
similarity_threshs = np.linspace(0, 1, 21)
for thresh in similarity_threshs:
    ident = apply_similarity_thresh(ident1, thresh)
    other_ident = apply_similarity_thresh(ident2, thresh)
    hallucinations = get_hallucinations(ident, other_ident)
    hallucination_metrics = get_hallucination_metrics(ident, hallucinations)
    results.loc[len(results)] = [thresh, hallucination_metrics['precision'], hallucination_metrics['recall'], hallucination_metrics['f1'], hallucination_metrics['custom']]
```

```python

def get_hallucination_rate(ident, other_ident, thresh):
    ident.replace(np.nan, '', inplace=True)
    other_ident.replace(np.nan, '', inplace=True)
    
    ident_starts = set()
    for i in range(len(ident)):
        if ident.loc[i, 'speaker'] != 'unknown':
            ident_starts.add(ident.loc[i, 'start'])

    other_ident_starts = set()
    other_ident_unknown_lens = []
    for i in range(len(other_ident)):
        if other_ident.loc[i, 'speaker'] != 'unknown':
            other_ident_starts.add(other_ident.loc[i, 'start'])
        else:
            transcription = other_ident.loc[i, 'transcription']
            if transcription == np.nan:
                transcription = ''
            other_ident_unknown_lens.append(len(transcription))
    other_ident_unknown_avg_len = np.mean(other_ident_unknown_lens)
    
    num_hallucinations = 0
    for other_ident_start in other_ident_starts:
        if other_ident_start not in ident_starts:
            num_hallucinations += 1
    
    if len(other_ident_starts) == 0:
        return 1 if len(ident_starts) == 0 else 0
    print()
    print(f"Ident: {len(ident) - len(ident_starts)},\tOther ident: {len(other_ident) - len(other_ident_starts)}")
    hallucination_rate = num_hallucinations / len(other_ident_starts)
    return hallucination_rate

similarity_threshs = np.linspace(0, 1, 21)
for thresh in similarity_threshs:
    ident = apply_similarity_thresh(ident1, thresh)
    other_ident = apply_similarity_thresh(ident2, thresh)
    hallucination_rate = get_hallucination_rate(ident, other_ident, thresh)
    print(f"Similarity threshold: {thresh:.2f},\tHallucination rate: {hallucination_rate:.2f}")
```
